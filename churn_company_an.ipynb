```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv", encoding = "ISO-8859-1")
df.head()
```

```python
df.replace("No internet service", "No", inplace=True)
df.replace("No phone service", "No", inplace=True)
```

```python
df.head()
```

```python
#just shown distribution of numerical columns
import seaborn as sns

numerical_columns = df.select_dtypes(include=['number']).columns
df[numerical_columns].hist(bins=15, figsize=(15, 6), layout=(2, 4))
plt.show()
```

```python
# shows that how Relationship between tenure and MonthlyCharges
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tenure', y='MonthlyCharges', hue='Churn', data=df)
plt.title('Tenure vs MonthlyCharges')
plt.show()

```

```python
# Correlation matrix
corr_matrix = df[numerical_columns].corr()
print(corr_matrix)

# Heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# Churn rate by contract type
plt.figure(figsize=(10, 6))
sns.countplot(x='Contract', hue='Churn', data=df)
plt.title('Churn Rate by Contract Type')
plt.show()

# Monthly Charges by Payment Method
plt.figure(figsize=(10, 6))
sns.boxplot(x='PaymentMethod', y='MonthlyCharges', hue='Churn', data=df)
plt.title('Monthly Charges by Payment Method')
plt.xticks(rotation=45)
plt.show()
```

```python
Q1 = df['MonthlyCharges'].quantile(0.25)
Q3 = df['MonthlyCharges'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers_iqr = df[(df['MonthlyCharges'] < lower_bound) | (df['MonthlyCharges'] > upper_bound)]



# Replace outliers with the median value
median_value = df['MonthlyCharges'].median()
df['MonthlyCharges'] = df['MonthlyCharges'].apply(lambda x: median_value if (x < lower_bound or x > upper_bound) else x)

print(f"Data after replacing outliers with median:\n{df}")
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
Q1 = df['TotalCharges'].quantile(0.25)
Q3 = df['TotalCharges'].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers_iqr = df[(df['TotalCharges'] < lower_bound) | (df['TotalCharges'] > upper_bound)]
print(f"Outliers using IQR:\n{outliers_iqr}")

# Replace outliers with the median value
median_value = df['TotalCharges'].median()
df['TotalCharges'] = df['TotalCharges'].apply(lambda x: median_value if (x < lower_bound or x > upper_bound) else x)

print(f"Data after replacing outliers with median:\n{df}")
```
